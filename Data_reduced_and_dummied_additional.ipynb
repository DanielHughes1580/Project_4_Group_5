## Processing

```python
import pandas as pd

#  Import and read the csv file.
df=pd.read_csv("https://raw.githubusercontent.com/DanielHughes1580/Project_4_Group_5/main/Data_Cleaned/data_clean.csv",index_col=0)
df.head()
```

```python
# Drop column ID.
df = df.drop(columns=['ID'])
df.head()
```

```python
# Determine the number of unique values in each column.
df.nunique()
```

```python
# Drop "owns a mobile" column as everone seems to have a mobile
df = df.drop(columns=['Owns a Mobile'])
df.head()
```

```python
# Look at data types
df.dtypes
```

```python
df["Income Category"].value_counts()
```

```python
# Remove Income Category of Student as too few instances
df = df.loc[(df["Income Category"] != "Student")]
df["Income Category"].value_counts()
```

```python
# Convert categorical data and strings to dummies
df = pd.get_dummies(df, columns=["Gender","Owns a Car","Owns Property","Income Category","Education Level","Marital Status","Houseing Type","Occupation"])
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Drop one of the two columns where there is a binary choice
df = df.drop(columns=['Owns a Car_N','Gender_M','Owns Property_N'])
```

```python
df.head()
```

```python
df.to_csv('../Data_Cleaned/dummied_data.csv')  
```
