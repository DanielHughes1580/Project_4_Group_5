```python
#  Import and read the csv file.
df=pd.read_csv("https://media.githubusercontent.com/media/DanielHughes1580/Project_4_Group_5/main/Data_Cleaned/dummied_data_add_clean.csv",index_col=0)
df.head()
```

```python
# Convert categorical labels to numerical labels
label_encoder = LabelEncoder()
df['status_encoded'] = label_encoder.fit_transform(df['Status'])

# Convert numerical labels to one-hot encoded format
encoded_status = to_categorical(df['status_encoded'])

# Split data into features and target
X = df.drop(columns=['Status', 'status_encoded'])  # Features
y = encoded_status  # Target

# Apply PCA
pca = PCA(n_components=10)  # Choose number of components as desired
X_pca = pca.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, random_state=78)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```
